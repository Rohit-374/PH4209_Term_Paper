```python
# Importing Useful Libraries

import numpy as np
import matplotlib.pyplot as plt
import time 
```

```python
# Importing the Data and doing the averaging

Data = []

for i in range(21):
    ST = []
    for j in range(21):
        D0 = np.loadtxt("Data_S"+str(i)+"_T"+str(j)+".csv", delimiter=",")
        G0 = (D0[0] + D0[1] + D0[2] + D0[3] + D0[4])/5
        F0 = np.sum(G0[-2001: -1])/len(G0[-2001: -1])
        ST.append(F0)
    Data.append(ST)
```

```python
# Creating an array of range of S and T value, and frequency of cooperator after averaging over the last 2000 iterations

S = np.linspace(-1, 1, 21)
T = np.linspace(0, 2, 21)
V = 1 - np.array(Data)
```

```python
# Creating Color plot

plt.pcolormesh(T, S, V, shading="gouraud", vmin=0, vmax=1, cmap='jet')
plt.colorbar()
  
# Adding details to the plot

plt.title( "Multiplex (L=5)" )
plt.xlabel('T value')
plt.ylabel('S value')

#Saving Plot
plt.savefig("Multiplex (L=5).pdf", format="pdf", bbox_inches="tight")

# Displaying the plot
plt.show()
```
