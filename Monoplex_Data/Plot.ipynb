```python
# Importing Useful Libraries

import numpy as np
import matplotlib.pyplot as plt
import time 
```

```python
# Importing the Data and doing the averaging

Data = []


for i in range(31):
    ST = []
    for j in range(31):
        D0 = np.loadtxt("Data_S"+str(i)+".csv", delimiter=",")
        G0 = D0[j]
        F0 = np.sum(G0[-501: -1])/len(G0[-501: -1])
        ST.append(F0)
    Data.append(ST)
        
```

```python
# Creating an array of range of S and T value, and frequency of cooperator after averaging over the last 500 iterations

S = np.linspace(-1, 1, 31)
T = np.linspace(0, 2, 31)
V = 1 - np.array(Data)
```

```python
# Creating Color plot

plt.pcolormesh(T, S, V, shading="gouraud", vmin=0, vmax=1, cmap='jet')
plt.colorbar()
  
# Adding details to the plot

plt.title( "Monoplex (L=1)" )
plt.xlabel('T value')
plt.ylabel('S value')

#Saving Plot
plt.savefig("Monoplex C.pdf", format="pdf", bbox_inches="tight")

# Displaying the plot
plt.show()
```
